```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.genfromtxt("CAPM test.csv",delimiter=",")
#points
#提取points中的两列数据，分别作为x,y
x=points[:,1];
y=points[:,0];

#用plt画出散点图
plt.scatter(x,y)
plt.show()
```

```python
# 损失函数是系数的函数，另外还要传入数据的x,y
def compute_cost(w,b,points):
    total_cost=0
    M =len(points)
    for i in range(M):
        x=points[i,1]
        y=points[i,0]
        total_cost += (y-w*x-b)**2
    return total_cost/M #一除都是浮点 两个除号是地板除，整型。 如 3 // 4
```

```python
#定义核心拟合函数
def fit(points):
    M = len(points)
    x_bar=np.mean(points[:,1])
    sum_yx= 0
    sum_x2=0
    sum_delta =0
    for i in range(M):
        x=points[i,1]
        y=points[i,0]
        sum_yx += y*(x-x_bar)
        sum_x2 += x**2
    #根据公式计算w
    w = sum_yx/(sum_x2-M*(x_bar**2))
    
    for i in range(M):
        x=points[i,1]
        y=points[i,0] 
        sum_delta += (y-w*x)
    b = sum_delta / M
    return w,b
```

```python
w,b =fit(points)
```

```python
print ("w is :",w)
print ("b is :",b)
cost = compute_cost(w,b,points)
print("cost is :" ,cost)
```
